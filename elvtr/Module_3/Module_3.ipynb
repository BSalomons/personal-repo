## Filtering

```python
df = pd.read_csv('titanic.csv')
passenger_count = len(df) 
print(f'Read in {passenger_count} rows.')
df.head()
```

```python
df.columns
```

```python
male_df = df[df['Sex'] == 'male']
male_count = len(male_df)
print(f'There were {male_count} males.')
male_df.head()
```

```python
female_df = df[df['Sex'] == 'female']
female_count = len(female_df)
print(f'There were {female_count} females.')
female_df.head()
```

```python
passenger_count == female_count + male_count # True if all passengers accounted for (by Sex)
```

```python
survived_count = len(df[df['Survived'] == 1])
perished_count = len(df[df['Survived'] == 0])
print(f'Of the {passenger_count} passengers, {survived_count} survived and {perished_count} perished.')
survival_ratio = survived_count / passenger_count
print(f'So {survival_ratio:.1%} survived.')
```

# Indexing

We want to index on a column (or columns) to help us retrieve information from the dataframe.

```python
df.head() # that left-most column is the index that Pandas gave us by default.
```

```python
indexed_df = df.set_index('PassengerId')
indexed_df.head()
```

```python
two_index_df = df.set_index(['Sex', 'PassengerId'])
two_index_df.head()
```

```python
# Although we changed the index, iloc will give the same results.
mini_df = indexed_df.iloc[[2, 4, 6, 8]] # Need two levels of brackets
mini_df # Why? The rows are still the same.
```
