```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr 
import datetime as dt
import matplotlib.pyplot as plt

NOW = dt.datetime.now()


class Input:
    def __init__(self):
        self.tickers = [str(x) for x in input('Enter Stock Tickers Separated by a Space: ').upper().split()]
        self.market = str(input('Enter an index to use: ').upper())
        self.tickers.insert(0, self.market)
        
        while True:
            try:
                self.start_date = dt.datetime.strptime(input('Enter a start date (YYYY-MM-DD): '), '%Y-%m-%d')
                if self.start_date >= NOW:
                    print('Date cannot be in the future')
                    continue
            
                while True:
                    try:
                        self.end_date = dt.datetime.strptime(input('Enter a end date (YYYY-MM-DD): '), '%Y-%m-%d')
                        if self.end_date <= self.start_date:
                            print('End Date must be later than Start Date')
                            continue
                        elif self.end_date >= NOW:
                            print('Date cannot be in the future')
                            continue
                    except ValueError:
                        print('Wrong format')
                        continue
                    else:
                        break
            except ValueError:
                print('Wrong format')
                continue
            else:
                break

    

class RiskPortfolio(Input):
    def __init__(self):
        super().__init__()


        self.df = None
        self.tickers_np = []
        self.market_np = []
        print("init called")
    
            
    def read_yahoo(self):
        yahoo = pdr.get_data_yahoo(self.tickers, self.start_date, self.end_date, interval="m")
        self.df = yahoo['Adj Close'].copy()
        print("df created")
               
"""
    def to_numpy(self):
        
        for stock in self.ticker:   
            close = self.np_array.append(self.df[stock])
        
        close = np.asarray(self.np_array)
        shift = np.roll(self.np_array, 1, axis=1)
        
        for x in range(0,len(shift)):
            shift[x,0] = np.nan
        
        ln_ret = np.log(close/shift)
        
        self.np_array = ln_ret[:,1:]
        self.np_index = ln_ret[0,1:]
        
        print('np_array created')        
"""        
        
            
input_data = RiskPortfolio()
```
